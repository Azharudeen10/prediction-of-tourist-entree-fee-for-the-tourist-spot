```python
import pandas as pd
import numpy as np
df = pd.read_csv('Top Indian Places to Visit.csv')
```

```python
df
```

```python
newdf = df[df['Entrance Fee in INR'] != 0]
```

```python
newdf
```

```python
x = newdf[['Type','Google review rating','Number of google review in lakhs']]
x
```

```python
y = newdf['Entrance Fee in INR']
y
```

```python
x['Type'].nunique()
```

```python
x = pd.get_dummies(x,columns =['Type'],dtype = 'int')
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
from sklearn.linear_model import LinearRegression
reg1 = LinearRegression()
reg1.fit(x_test,y_test)
```

```python
reg1.score(x_test,y_test)
```

```python
reg1.score(x_train,y_train)
```

```python
from sklearn.linear_model import Lasso
reg3 = Lasso(alpha=25, max_iter=100, tol=0.1)
reg3.fit(x_train,y_train)
reg3.score(x_test,y_test)
```

```python
reg3.score(x_train,y_train)
```

```python
from sklearn.linear_model import Ridge
reg4 = Ridge(alpha=25, max_iter=100, tol=0.1)
reg4.fit(x_train,y_train)
reg4.score(x_test,y_test)
```

```python
reg4.score(x_train,y_train)
```
